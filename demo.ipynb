## Demo Notebook for Human Skeleton Tracking using Deep Learning

### Import pre-requisites

```python
from PosefromVid import posefromvid
from video_writer import writeSkeleton
import inference_3d
from vis3d import animate3d
import numpy as np
import pickle
import warnings
warnings.filterwarnings("ignore")
```

### Initiate input video file

```python
videofile = 'sample/shortened_test.mp4' # Enter path to your video file
```

### Generate 2D pose file

```python
pose2d = posefromvid(videofile)
```

### Write tracked skeleton on video

In this stage, the tracked skeleton is overlayed on the video. The result is stored in 'output2d' folder in the working directory.

```python
writeSkeleton(videofile, pose2d)
```

### Generate 3D detections

The best performing model based on the evaluation results in the report has been initialised below. Different model can be selected from 3D_model folder

```python
model_json = '3D_model/multistage_best/model_multistage_dropout.json'
model_weight = '3D_model/multistage_best/model_multistage_weight_dropout.h5'
```

```python
infer3d = inference_3d.Inference(model_json, model_weight, pose2d)
```

```python
kp3d = infer3d.inference()
```

```python
kp3d = np.array(kp3d)
```

```python
with open('prediction.pickle', 'wb') as handle:
    pickle.dump(kp3d, handle, protocol=pickle.HIGHEST_PROTOCOL)
handle.close()
```
